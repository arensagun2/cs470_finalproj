```python
import torch
import torchvision
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Define your transformations (customize as needed)
transform = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.ToTensor(),
])
```

```python
# Define a function to check if a file is a valid image file
def is_valid_file(path):
    return path.endswith('.jpg') or path.endswith('.jpeg') or path.endswith('.png') or path.endswith('.ppm') or path.endswith('.bmp') or path.endswith('.pgm') or path.endswith('.tif') or path.endswith('.tiff') or path.endswith('.webp')
```

```python
# Create a dataset
dataset = datasets.ImageFolder(root='C:/Users/jlazz/OneDrive/Desktop/CS470/CS470_Term_Project/470Data', transform=transform, is_valid_file=is_valid_file)
```

```python
# Initialize DataLoader
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)

for images, labels in dataloader:
    print(f"Batch images shape: {images.shape}")
    print(f"Batch labels shape: {labels.shape}")
    # Break after the first batch to only process it
    break
```

```python
# Get one batch of images and labels
images, labels = next(iter(dataloader))
```

```python
# Make a grid from batch
grid = torchvision.utils.make_grid(images)
```

```python
# Convert tensor to numpy for visualization
numpy_grid = grid.numpy()
```

```python
# Transpose the numpy_grid
numpy_grid = np.transpose(numpy_grid, (1, 2, 0))
```

```python
# Plot the grid
plt.figure(figsize=(10,10))
plt.imshow(numpy_grid)
plt.show()
```
